```python
# 

temp_d = char['Death'].copy()
temp_d = temp_d.str.rstrip().str.rstrip('s')
temp_d = temp_d.str[-4:]
temp_d.fillna('0',inplace=True)
temp_d[~temp_d.str.isdigit()]=0
temp_d = temp_d.astype('int32')

temp = char[['Birth']].copy()
temp['Death'] = temp_d
temp = temp.query('(Death >= 1993)|(Death == 0)')
temp = temp['Birth']

temp = temp.str.rstrip().str.rstrip('s')
temp.fillna('no_inf',inplace=True)
temp = temp.str[-4:]
temp.drop(temp[~temp.str.isdigit()].index,inplace=True)
temp = temp.astype('int32')
temp = np.abs(temp - 1993)
char['Age'] = temp
```

```python
#    
spel.drop(
    spel.query(
        '(Incantation=="Unknown")|(Incantation=="None")')
    .index,inplace=True) 

#    
sp = spel['Incantation'].dropna().unique().copy() 

#   
spels = pd.merge(
    dial,char2[['Character Name','Character ID']],how='left') 

def sp_count(x): 
    for i in sp:
        if i+'.' in x or i+'!' in x:
            return i
        
spel_c = spels.Dialogue.map(sp_count)        
spel_count = spel_c.value_counts().head(10)

#
plt.figure(figsize=(15,4))
x = spel_count.index
y = spel_count.values

plt.ylabel('',labelpad = 10.0, fontdict = {'family':'serif','color':'maroon','size':15})
plt.xlabel(' ',labelpad = 10.0, fontdict = {'family':'serif','color':'maroon','size':15})
plt.title(' 10   ',pad = 20.0, fontdict = {'family':'serif','size':15})

plt.bar(x,y,color='sienna',width=0.4)
plt.show()
```
