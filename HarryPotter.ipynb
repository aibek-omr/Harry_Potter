## Узнаем возраст волшебников (в счет берутся только живые волшебники)

```python
#Расчет возраста

temp_d = char['Death'].copy()
temp_d = temp_d.str.rstrip().str.rstrip('s')
temp_d = temp_d.str[-4:]
temp_d.fillna('0',inplace=True)
temp_d[~temp_d.str.isdigit()]=0
temp_d = temp_d.astype('int32')

temp = char[['Birth']].copy()
temp['Death'] = temp_d
temp = temp.query('(Death >= 1993)|(Death == 0)')
temp = temp['Birth']

temp = temp.str.rstrip().str.rstrip('s')
temp.fillna('no_inf',inplace=True)
temp = temp.str[-4:]
temp.drop(temp[~temp.str.isdigit()].index,inplace=True)
temp = temp.astype('int32')
temp = np.abs(temp - 1993)
char['Age'] = temp
```

Какие заклинания чаще всего использовались

```python
#убираем неизвестные залинания без названия
spel.drop(
    spel.query(
        '(Incantation=="Unknown")|(Incantation=="None")')
    .index,inplace=True) 

#удаляем пропуски и выбираем уникальные
sp = spel['Incantation'].dropna().unique().copy() 

#добавляем в диалоги имена
spels = pd.merge(
    dial,char2[['Character Name','Character ID']],how='left') 

def sp_count(x): 
    for i in sp:
        if i+'.' in x or i+'!' in x:
            return i
        
spel_c = spels.Dialogue.map(sp_count)        
spel_count = spel_c.value_counts().head(10)

#График
plt.figure(figsize=(15,4))
x = spel_count.index
y = spel_count.values

plt.ylabel('Количество',labelpad = 10.0, fontdict = {'family':'serif','color':'maroon','size':15})
plt.xlabel('Название заклинания',labelpad = 10.0, fontdict = {'family':'serif','color':'maroon','size':15})
plt.title('Топ 10 часто используемых заклинаний',pad = 20.0, fontdict = {'family':'serif','size':15})

plt.bar(x,y,color='sienna',width=0.4)
plt.show()
```

Использовали ли главные герои (Гарри, Рон и Гермиона) непростительные заклинания (Avada Kedavra, Imperio, Crucio).

```python
spels['Spells'] = spel_c
spels = spels[spels.Spells.notna()]
fbdn = spel[spel['Incantation'].isin(['Avada Kedavra','Imperio','Crucio'])]['Incantation']
main_char = spels[spels['Character Name'].str.contains('Harry|Ron|Hermione')]
main_char[main_char.Spells.isin(fbdn)].iloc[:,-3:]
```

Как видно из таблицы выше Гарри использовал 2 запретных заклинания Crucio и Imperio. Рон так же использовал Imperio. А Гермиона не использовала запретные заклинания.
